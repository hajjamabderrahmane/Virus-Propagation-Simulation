```python
# === 1. Importations ===
import numpy as np
import matplotlib.pyplot as plt
import sympy as sp

# === 2. Données réelles de Rétablis (R) ===
R_vals = [
     0, 3, 820, 1755, 1695, 1618, 1182, 1312, 396, 361, 238, 191,
    327, 128, 114, 111, 106, 157, 117, 190, 127, 107, 100, 104,
    114, 124, 160, 170, 243, 491, 341, 284, 225, 368, 308, 434,
    744, 790, 1020, 1113, 1558, 1918, 2066, 2599, 3181, 8831, 2522,
    2519, 2515, 1255, 1572, 1950, 1721
]

x_vals = list(range(len(R_vals)))  # Semaine 0 à 52
x = sp.Symbol('x')

# === 3. Interpolation de Newton ===
def newton_interpolation(x_vals, y_vals, var):
    n = len(x_vals)
    coef = y_vals.copy()
    for j in range(1, n):
        for i in range(n - 1, j - 1, -1):
            coef[i] = (coef[i] - coef[i - 1]) / (x_vals[i] - x_vals[i - j])
    poly = coef[0]
    for i in range(1, n):
        term = coef[i]
        for j in range(i):
            term *= (var - x_vals[j])
        poly += term
    return sp.simplify(poly)

# === 4. Calcul et affichage du polynôme ===
P = newton_interpolation(x_vals, R_vals, x)

print("Polynôme de Newton (R, degré 52) :")
print(P)
```

```python
import numpy as np
from datetime import datetime, timedelta

# Données de R (rétablis)
R_vals = [
    0, 3, 820, 1755, 1695, 1618, 1182, 1312, 396, 361, 238, 191, 327, 128, 114,
    111, 106, 157, 117, 190, 127, 107, 100, 104, 114, 124, 160, 170, 243, 491,
    341, 284, 225, 368, 308, 434, 714, 790, 1020, 1713, 1558, 1918, 2066, 2599,
    3121, 2831, 2522, 2519, 2515, 1255, 1572, 1950, 1721
]

# Paramètres
start_date = datetime(2020, 3, 1)
degree = 7
start_t = len(R_vals)
weeks = 10

# Créer les t (en jours) correspondants aux données R
t_vals = [7 * i for i in range(len(R_vals))]

# Ajustement polynôme de degré 7
coeffs = np.polyfit(t_vals, R_vals, degree)

# Fonction R(t)
def R(t):
    return np.polyval(coeffs, t)

# Générer prédictions futures
print("Date\t\t| t\t| R(t) prédit")
print("----------------|-------|------------")
for k in range(weeks):
    t = 7 * (start_t + k)  # t en jours (pas en semaines !)
    date = start_date + timedelta(days=t)
    R_pred = max(round(R(t)), 0)
    print(f"{date.strftime('%d/%m/%Y')}\t| {t}\t| {R_pred}")



```
