```python
import pandas as pd
import numpy as np
from datetime import datetime

# Données New York
data = {
    'Date': ['1/3/2020', '8/3/2020', '15/3/2020', '22/3/2020', '29/3/2020', '5/4/2020', '12/4/2020', '19/4/2020',
             '26/4/2020', '3/5/2020', '10/5/2020', '17/5/2020', '24/5/2020', '31/5/2020', '7/6/2020', '14/6/2020',
             '21/6/2020', '28/6/2020', '5/7/2020', '12/7/2020', '19/7/2020', '26/7/2020', '2/8/2020', '9/8/2020',
             '16/8/2020', '23/8/2020', '30/8/2020', '6/9/2020', '13/9/2020', '20/9/2020', '27/9/2020', '4/10/2020',
             '11/10/2020', '18/10/2020', '25/10/2020', '1/11/2020', '8/11/2020', '15/11/2020', '22/11/2020',
             '29/11/2020', '6/12/2020', '13/12/2020', '20/12/2020', '27/12/2020', '3/1/2021', '10/1/2021',
             '17/1/2021', '24/1/2021', '31/1/2021', '7/2/2021', '14/2/2021', '21/2/2021', '28/2/2021'],
    'I': [0, 21, 1034, 2580, 3560, 3782, 2887, 2342, 1003, 787, 457, 360, 467, 230, 197, 180, 172, 201, 155, 223,
          175, 140, 135, 133, 140, 150, 186, 207, 263, 206, 382, 335, 281, 412, 362, 501, 788, 896, 1148, 1891,
          1753, 2177, 2346, 2888, 3464, 3197, 2919, 2927, 2894, 1645, 1898, 2286, 2037],
    'R': [0, 3, 820, 1755, 1695, 1618, 1182, 1312, 396, 361, 238, 191, 327, 128, 114, 111, 106, 157, 117, 190,
          127, 107, 100, 104, 114, 124, 160, 170, 243, 491, 341, 284, 225, 368, 308, 434, 744, 790, 1020, 1113,
          1558, 1918, 2066, 2599, 3181, 8831, 2522, 2519, 2515, 1255, 1572, 1950, 1721],
    'D': [0, 0, 4, 52, 356, 758, 754, 494, 300, 210, 114, 89, 54, 42, 29, 22, 22, 12, 9, 7,
          10, 8, 2, 3, 5, 3, 6, 6, 3, 5, 4, 7, 7, 4, 10, 12, 9, 14, 12, 15,
          22, 32, 34, 48, 43, 56, 59, 74, 76, 80, 83, 62, 72]
}

# Préparation des données
df = pd.DataFrame(data)
df['Date'] = pd.to_datetime(df['Date'], format='%d/%m/%Y')
df = df.sort_values('Date').reset_index(drop=True)
df['N'] = 8_250_000  # Population totale de New York
df['S'] = df['N'] - df['I'] - df['R'] - df['D']
df['delta_days'] = df['Date'].diff().dt.days.fillna(7)

# Calcul robuste des paramètres a et b
def safe_divide(numerator, denominator):
    return np.where(denominator != 0, numerator/denominator, np.nan)

df['a'] = safe_divide(df['R'].diff(), df['I'].shift(1) * df['delta_days'])
df['b'] = safe_divide(df['D'].diff(), df['I'].shift(1) * df['delta_days'])

# Paramètres moyens (en excluant les valeurs aberrantes)
valid_a = df['a'][(df['a'] > 0) & (df['a'] < 1)].dropna()
valid_b = df['b'][(df['b'] > 0) & (df['b'] < 1)].dropna()
a = valid_a.mean()
b = valid_b.mean()

# Calcul de dI/dt observé
df['dI_dt_obs'] = df['I'].diff() / df['delta_days']

# Estimation robuste de r pour chaque point
def estimate_r(row):
    if row['I'] == 0 or row['S'] == 0 or np.isnan(row['dI_dt_obs']):
        return np.nan
    numerator = row['dI_dt_obs'] + (a + b) * row['I']
    denominator = row['S'] * row['I']
    return numerator / denominator if denominator != 0 else np.nan

df['r'] = df.apply(estimate_r, axis=1)

df['r'] = df['r'].fillna(df['r'].rolling(3, min_periods=1).mean())

# Identification du point de décroissance
df['dI_dt_model'] = df['r'] * df['S'] * df['I'] - (a + b) * df['I']
decline_start = df[df['dI_dt_model'] < 0].iloc[0]

print("Paramètres moyens calculés:")
print(f"a (taux de guérison) = {a:.4f}")
print(f"b (taux de mortalité) = {b:.4f}")
print(f"r (taux de contact) à la date de décroissance = {decline_start['r']:.4e}\n")

print("Début de la décroissance:")
print(f"   - Date: {decline_start['Date'].strftime('%d/%m/%Y')}")
print(f"   - Cas infectés (I): {int(decline_start['I'])}")
print(f"   - Individus susceptibles (S): {int(decline_start['S'])}")
print(f"   - dI/dt: {decline_start['dI_dt_model']:.2f} cas/jour (première valeur négative)")
```
